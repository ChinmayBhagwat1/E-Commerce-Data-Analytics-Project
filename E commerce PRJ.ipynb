```python
df['Discounted Price'] = df['Price'] * (1 - df['Discount'] * 0.01)
df['Price After Tax'] = df['Discounted Price'] * (1 + df['Tax Rate'] * 0.01)
```

```python
df['Price After Tax'] = pd.to_numeric(df['Price After Tax'], errors = 'coerce')
df['Shipping Cost'] = pd.to_numeric(df['Shipping Cost'], errors = 'coerce')
```

```python
df['Final Price'] = df['Price After Tax'] + df['Shipping Cost']
```

```python
df['Profit Margin'] = ((df['Final Price'] - df['Price']) / df['Final Price']) * 100
```

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
df.columns
```
